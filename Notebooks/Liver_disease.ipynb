```python
import pandas as pd
```

```python
df = pd.read_csv(r'Data\liver.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Diagnosis', axis=1)
y = df['Diagnosis']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr_train_pred = lr.predict(X_train)
lr_train_acc = accuracy_score(lr_train_pred, y_train)
print(lr_train_acc)
```

```python
lr_test_pred = lr.predict(X_test)
lr_test_acc = accuracy_score(lr_test_pred, y_test)
print(lr_test_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
rfc_train_pred = rfc.predict(X_train)
rfc_train_acc = accuracy_score(rfc_train_pred, y_train)
print(rfc_train_acc)
```

```python
rfc_test_pred = rfc.predict(X_test)
rfc_test_acc = accuracy_score(rfc_test_pred, y_test)
print(rfc_test_acc)
```

```python
import pickle
```

```python
filename = 'liver_disease_model.sav'
pickle.dump(rfc, open(filename, 'wb'))
```
