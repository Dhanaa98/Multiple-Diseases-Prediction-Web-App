```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\PC User\Desktop\Me\DS\_PROJECTS\MLearning\Projects\PUBLISHED\Multiple Diseases Prediction WebApp\Datasets\diabetes_prediction_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['smoking_history'] = df['smoking_history'].replace({'never':0, 'No Info':0, 'former':0, 'not current':0, 'current':1, 'ever':1})
df['gender'] = df['gender'].replace({'Female':0, 'Male':1, 'Other':2})
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.sample(20000)
```

```python
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr_train_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr_train_acc = accuracy_score(lr_train_pred, y_train)
print(lr_train_acc)
```

```python
lr_test_pred = lr.predict(X_test)
```

```python
lr_test_acc = accuracy_score(lr_test_pred, y_test)
print(lr_test_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_train_pred = rfc.predict(X_train)
```

```python
rfc_train_acc = accuracy_score(rfc_train_pred, y_train)
print(rfc_train_acc)
```

```python
rfc_test_pred = rfc.predict(X_test)
rfc_test_acc = accuracy_score(rfc_test_pred, y_test)
print(rfc_test_acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=50)
```

```python
dtc.fit(X_train, y_train)
```

```python
dtc_train_pred = dtc.predict(X_train)
dtc_train_acc = accuracy_score(dtc_train_pred, y_train)
print(dtc_train_acc)
```

```python
dtc_test_pred = dtc.predict(X_test)
dtc_test_acc = accuracy_score(dtc_test_pred, y_test)
print(dtc_test_acc)
```
