```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'Data\diabetes.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr_train_pred = lr.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr_train_acc = accuracy_score(lr_train_pred, y_train)
print(lr_train_acc)
```

```python
lr_test_pred = lr.predict(X_test)
```

```python
lr_test_acc = accuracy_score(lr_test_pred, y_test)
print(lr_test_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_train_pred = rfc.predict(X_train)
```

```python
rfc_train_acc = accuracy_score(rfc_train_pred, y_train)
print(rfc_train_acc)
```

```python
rfc_test_pred = rfc.predict(X_test)
rfc_test_acc = accuracy_score(rfc_test_pred, y_test)
print(rfc_test_acc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_depth=50)
```

```python
dtc.fit(X_train, y_train)
```

```python
dtc_train_pred = dtc.predict(X_train)
dtc_train_acc = accuracy_score(dtc_train_pred, y_train)
print(dtc_train_acc)
```

```python
dtc_test_pred = dtc.predict(X_test)
dtc_test_acc = accuracy_score(dtc_test_pred, y_test)
print(dtc_test_acc)
```

```python
from sklearn import svm
```

```python
svm = svm.SVC(kernel = 'linear')
```

```python
svm.fit(X_train, y_train)
```

```python
svm_train_pred = svm.predict(X_train)
svm_train_acc = accuracy_score(svm_train_pred, y_train)
print(svm_train_acc)
```

```python
svm_test_pred = svm.predict(X_test)
svm_test_acc = accuracy_score(y_test, svm_test_pred)
print(svm_test_acc)
```
